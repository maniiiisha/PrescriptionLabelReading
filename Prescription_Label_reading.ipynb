### **Problem Statement:**
Support elderly or vulnerable patients should be a focus for many businesses This is especially true for those in the health care sector.
```python
from paddleocr import PaddleOCR, draw_ocr
from matplotlib import pyplot as plt
import cv2
import os
```

### **Instantiating the model and performing detection**

```python
ocr = PaddleOCR(lang ='en')
```

```python
img_path = os.path.join('.','drug1.jpg')
```

```python
result = ocr.ocr(img_path)
```

```python
result[0]
```

### **Visualizing the result**

```python
boxes = [res[0][0] for res in result[0]]
txts = [res[1][0] for res in result[0]]
scores = [res[1][1] for res in result[0]]
```

```python
print(boxes, "\n", "------------------------------------------------------------------")
print(txts, "\n", "------------------------------------------------------------------")
print(scores, "\n", "------------------------------------------------------------------")
```

```python
font_path = os.path.join('/content/PaddleOCR/doc/fonts/', 'latin.ttf')
```

```python
img = cv2.imread(img_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize= (15,15))
annotation = draw_ocr(img, boxes, txts, scores, font_path = font_path)
plt.imshow(annotation)
```

```python
txt_file = open("result.txt", "r")
t2s = txt_file.read()
tt = gTTS(text=t2s, lang='en', slow=False)
tt.save("txt2speech.mp3")
```

```python
from pdfminer.high_level import extract_text
import gradio

def pdf_to_text(file_obj):
   text = extract_text(file_obj.name)
   myobj = gTTS(text=t2s, lang='en', slow=False)
   myobj.save("result.wav")
   return 'result.wav'

iface = gradio.Interface(
   fn = pdf_to_text,
   inputs = 'file',
   outputs = 'audio'
   )
iface.launch()
```
